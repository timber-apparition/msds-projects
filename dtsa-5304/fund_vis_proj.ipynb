```python
# Source data can be found at https://www.kaggle.com/datasets/jackdaoud/esports-earnings-for-players-teams-by-game

import pandas as pd
import altair as alt

data= pd.read_csv("highest_earning_players.csv")
data= data.groupby(['CountryCode','Game'])['TotalUSDPrize'].sum().reset_index()
data.head()
```

```python
alt.Chart(data).mark_bar().encode(x="CountryCode", y="TotalUSDPrize")
```

```python
alt.Chart(data).mark_line().encode(x="CountryCode", y="TotalUSDPrize")
```

```python
alt.Chart(data).mark_circle().encode(x="CountryCode", y="TotalUSDPrize")
```

```python
alt.Chart(data).mark_rect().encode(x="CountryCode", y="TotalUSDPrize",
                                     color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                     tooltip=["Game", "TotalUSDPrize"])
```

```python
alt.Chart(data).mark_circle().encode(x="CountryCode", y="TotalUSDPrize",
                                     size=("TotalUSDPrize"),
                                     color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                     tooltip=["Game", "TotalUSDPrize"])
review_iteration_1 = alt.Chart(data).mark_circle().encode(x="CountryCode", y="TotalUSDPrize",
                                                          size=("TotalUSDPrize"),
                                                          color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                                          tooltip=["Game", "TotalUSDPrize"])

review_iteration_1.save('review_iteration_1.html', embed_options={'renderer':'svg'})
```

```python
""" Reducing to top 20 after review feedback, stated vis was too wide """

data10 = data.nlargest(n=31, columns=['TotalUSDPrize'])

#print(data10)

alt.Chart(data10).mark_circle().encode(x="CountryCode", y="TotalUSDPrize",
                                     size=("TotalUSDPrize"),
                                     color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                     tooltip=["Game", "TotalUSDPrize"])

review_iteration_2 = alt.Chart(data10).mark_circle().encode(x="CountryCode", y="TotalUSDPrize",
                                                            size=("TotalUSDPrize"),
                                                            color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                                            tooltip=["Game", "TotalUSDPrize"])

review_iteration_2.save('review_iteration_2.html', embed_options={'renderer':'svg'})
```

```python
alt.Chart(data10).mark_circle().encode(x="CountryCode", y="TotalUSDPrize",
                                       size=("TotalUSDPrize"),
                                       color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                       tooltip=["Game", "TotalUSDPrize"])
```

```python
"""Adding Selection to to make single games searchable"""

selection = alt.selection_multi(fields=['Game'], bind='legend')

alt.Chart(data10).mark_circle().encode(x="CountryCode", y="TotalUSDPrize",
                                       size=("TotalUSDPrize"),
                                       color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                       opacity=alt.condition(selection, alt.value(1), alt.value(0.2)),
                                       tooltip=["Game", "TotalUSDPrize"]).add_selection(selection)

review_iteration_3 =alt.Chart(data10).mark_circle().encode(x="CountryCode", y="TotalUSDPrize",
                                       size=("TotalUSDPrize"),
                                       color=alt.Color("Game", scale=alt.Scale(scheme='turbo')),
                                       opacity=alt.condition(selection, alt.value(1), alt.value(0.2)),
                                       tooltip=["Game", "TotalUSDPrize"]).add_selection(selection)
review_iteration_3.save('review_iteration_final.html', embed_options={'renderer':'svg'})
```
